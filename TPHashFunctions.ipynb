### Fonctions: 

```python
def read(nom):
    """fonction qui prend en paramètre le nom de fichier et renvoie:
    -contenu: la liste des mots du fichier
    -M: la taille du tableau de hachage """
    
    fs=open(nom, "r")
    contenu=fs.readlines()
    contenu=[mot.replace("\n", "") for mot in contenu]
    M=len(contenu)/0.3
    M=math.ceil(M)
    M=sympy.nextprime(M)
    return contenu,M
    
```

```python
def tableau_hachage(fonction,list_mot,M):
    """fonction qui prend en paramètre 3 arguments:
    -fonction: la fonction de hachage
    -list_mot: la liste des mots du fichier
    -M: la taille du tableau
    et renvoie:
    -tab_ha: la table de hachage """
    tab_ha=np.zeros(M)
    i=0
    for  j in range(len(list_mot)):
        i=fonction(list_mot[j],M)
        tab_ha[i]+=1
    return tab_ha
```

```python
def affichage (fonction,list_mot,M):
    """fonction qui prend en paramètre 3 arguments:
    -fonction: la fonction de hachage
    -list_mot: la liste des mots du fichier
    -M: la taille du tableau 
    et renvoie
    - L: la liste des mots avec lequels existent des collisions"""
    tab_ha=np.zeros(M)
    i=0
    L=[]
    for  j in range(len(list_mot)):
        i=fonction(list_mot[j],M)
        tab_ha[i]+=1
        if tab_ha[i]>1:
            L.append(list_mot[j])
    return L 
```

```python
#La performance 
def Performance(tab_ha):
    """fonction qui prend en paramètre la table de hachage et renvoie les 4 critères de performance"""
    m = len(tab_ha)
    sans_Acces = np.count_nonzero(tab_ha== 0)
    Max = np.max(tab_ha)
    Min = np.min(tab_ha[tab_ha != 0])
    nb_elements = np.count_nonzero(tab_ha)
    if nb_elements != 0:
        moy_1 = nb_elements / m
        moy_2 = np.sum(tab_ha) / nb_elements
    else:
        moy_1 = 0
        moy_2 = 0
    return sans_Acces, Min, Max, moy_1, moy_2

# Le nombre de collision
def collision(tab_ha):
    """ fonction qui prend en paramètre la table de hachage et renvoie le nombre de collision"""
    return sum(tab_ha[i]-1 for i in range(len(tab_ha)) if tab_ha[i]>1)
```

```python
# les différents fonctions de hachage: 
#La première fonction

def hash_Najlaa_Imane_1 (s,M):
    """la fonction prend en pramètre une chaine de caractère et la valeur M et renvoie l'index du mot dans la table hachage """
    g=31
    hash=0
    for i in range(0,len(s)) :
        hash = (hash*g + ord(s[i]))
    hash=hash%M
    return hash

#La deuxième fonction
def ascii_function(mot):
    """fonction prend en argument une chaine de caractère et renvoie la liste du code ASCII du mot """
    ascii= [ord(lettre)%97 for lettre in mot]
    return ascii

def hash_Najlaa_Imane_2(s,M):
    """ fonction de hachage qui se base sur la représentation dans la base 26"""
    ascii=ascii_function(s)
    ascii=ascii +[len(s)]
    hash= 0
    for i in range(len(ascii)):
        hash=hash+((ascii[i])*(26**(len(ascii)-i-1)))
    hash=hash%M
    return hash

# Amèlioration
def hash_Najlaa_Imane_3(s,M):
    """  Amélioration de  la fonction de hachage précédente en faisant une permutation si le code ASCII commence par 0   """
    ascii=ascii_function(s)
    if ascii[0]==0:
        i=0
        while ascii[i]==0 and i+1<len(s):
            i=i+1
        ascii=ascii[i:]+ascii[:i]
    hash=0 
    for i in range(len(ascii)):
        hash=hash+((ascii[i])*(26**(len(ascii)-i-1)))
    hash=hash%M
    return hash

#Amélioration 2
def hash_Najlaa_Imane_4(s,M):
    ascii=ascii_function(s)
    g=31
    ascii=ascii +[len(s)]
    hash= 0
    for i in range(0,len(s)) :
        hash = (hash*g + ord(s[i]))
    for i in range(len(ascii)):
        hash=hash+((ascii[i])*(26**(len(ascii)-i-1)))

    hash^= (hash >> 16)
    hash ^= (hash >> 9)
    hash += (hash << 15)
    hash += (hash << 3)
    hash=hash%M
    return hash
```

### Comparatif facteurs de multiplication aléatoires 

```python
def ASCII_g_ponderated (s ,M):
    #generer un nombre intègre 10 and 100
    x =random.randint(10, 100)
    # Generate a random prime 
    g = randprime(2**x, 2**(x+1))
    hash=0
    for i in range(0,len(s)) :
        hash =  (hash* g + ord(s[i]))
    hash=hash%M
    return hash,g
```

```python
# le tableau de hachage 
def tableau_hachage_1(fonction,list_mot,M):
    """fonction qui renvoie la table de hachage ainsi que le facteur de multiplication"""
    tab_ha=np.zeros(M)
    i=0
    for  j in range(len(list_mot)):
        i,g=fonction(list_mot[j],M)
        tab_ha[i]+=1
    return (tab_ha,g)
```

### Comparaison avec un autre groupe: 

```python
def hash_fonction1(word):
    hash = np.uint32(0)
    for c in word:
        hash += (ord(c) <<  5) | (ord(c) & hash)
        hash ^= (hash >> 6) + hash
    hash ^= (hash >> 8) | ~(hash << 12)
    hash = (hash << 3)
    return abs(hash)

def hash_fonction2(word):
    hash = np.uint32(0)
    for c in word:
        hash += ((ord(c)  << 2) & ~(ord(c) << 3))
        hash = (hash  << 3)
        hash -= (hash >> 5)
        hash ^= (hash >> 7)
    hash ^= (hash << 1) & ~(hash >> 17)
    hash = (hash >> 2) & ~(hash << 13)
    hash -= (hash >> 3)
    return abs(hash)
def nacho_diego_hachage(word, M):
    cle1 = hash_fonction1(word)
    cle2 = hash_fonction2(word)
    finalHash = (cle1 + cle2) % M
    return finalHash
```
